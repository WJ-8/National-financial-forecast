```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family'] = 'Microsoft YaHei'
```

```python
with open("train.txt","r") as f:
    data=f.readlines()
mae=[]
for i in data:
    mae.append(float(i.strip()))
plt.plot([i for i in range(10000)],mae,color="orange")
plt.box(False)
plt.grid(True)
plt.xlabel('迭代次数')  # X 轴标签
plt.ylabel('mae')  # Y 轴标签

plt.show()
```
